```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Generating a simulated dataset

np.random.seed(42)
mu, sigma = 0, 1 # mean and standard deviation
x = np.random.normal(mu, sigma, 100)
n = np.random.normal(mu, sigma, 100)
y = x - 2*x**2 + n
```

n = irreducible error(epsilon); number of datapoints in the dataset = 100; p = 2; 

y = x - 2x^2 + n

**b)**

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

The scatter plot shows the spread of x with respect to y and it is not a linear relationship. x is in the range -2 to 2.

**c) (i)**

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import LeaveOneOut, cross_val_score
```

```python
rgr = LinearRegression()
loocv = LeaveOneOut()
poly = PolynomialFeatures(1) 
polylin = poly.fit_transform(x.reshape(-1,1))
scores = cross_val_score(rgr, polylin, y, cv=loocv, scoring='neg_mean_squared_error')
print("Test MSE:",abs(scores.mean())) 
```

```python
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.scatter(x, y, label='Datapoints')
xp = np.linspace(-2, 2, 50)
plt.plot(xp ,p(xp), color='red', label='Linear Regression Line')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
```

**(f)**

```python
import statsmodels.formula.api as sm
```

```python
df = pd.DataFrame(x,y)
results = sm.ols('y ~ x', df).fit()
print(results.summary())
```

The coefficients are all statistically significant as they all have p value less than 0.001.

```python
x1 = x**2
data = np.c_[x,x1]
df1 = pd.DataFrame(data, y)
results = sm.ols('y ~ x+x1', df1).fit()
print(results.summary())
```

One of the coefficients seem to have a high p value and two of the coefficients have low p value.

```python
x2 = x**3
data = np.c_[x,x1,x2]
df2 = pd.DataFrame(data, y)
results = sm.ols('y ~ x+x1+x2', df2).fit()
print(results.summary())
```

The p value of the coefficients of x and x^2 are significant in the equation

```python
x3 = x**4
data = np.c_[x,x1,x2,x3]
df2 = pd.DataFrame(data, y)
results = sm.ols('y ~ x+x1+x2+x3', df2).fit()
print(results.summary())
```

The p value of the coefficients of x and x^2 are low. Therefore, they are significant in the equations

The results obtained through least squares method matches with the cross-validation. The quadratic polynomial has the highest R-squared value.
