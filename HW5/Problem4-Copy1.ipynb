```python
caravan['Purchase']= pd.Categorical(caravan['Purchase'],['Yes','No'])
caravan['Purchase']= caravan['Purchase'].cat.codes
print(caravan)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier
```

```python
y_train = caravan_train['Purchase']
X_train = caravan_train.drop('Purchase', axis = 1)

y_val = caravan_test['Purchase']
X_val = caravan_test.drop('Purchase', axis = 1)
```

```python
caravan_ip = caravan.drop('Purchase', axis = 1)
```

```python
regr = GradientBoostingRegressor(n_estimators=1000, learning_rate=0.01, max_depth=2, random_state=0).fit(X_train, y_train)
```

```python
imp_feature = regr.feature_importances_*100
set1 = pd.Series(imp_feature[0:15], index=caravan_ip.columns[0:15]).sort_values(inplace=False)
set1.T.plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
set2 = pd.Series(imp_feature[15:30], index=caravan_ip.columns[15:30]).sort_values(inplace=False)
set2.T.plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
set3 = pd.Series(imp_feature[30:45], index=caravan_ip.columns[30:45]).sort_values(inplace=False)
set3.T.plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
set4 = pd.Series(imp_feature[45:60], index=caravan_ip.columns[45:60]).sort_values(inplace=False)
set4.T.plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

```python
set5 = pd.Series(imp_feature[60:86], index=caravan_ip.columns[60:86]).sort_values(inplace=False)
set5.T.plot(kind='barh', color='r', )
plt.xlabel('Variable Importance')
plt.gca().legend_ = None
```

After performing regression, all the features were plotted on the basis of their importance, we can see that PPERSAUT and MOSTYPE are the most important feature.

**Logistic Regression**

```python
import sklearn.metrics as metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report, confusion_matrix
```

```python
clf1 = LogisticRegression(random_state=42)
clf1.fit(X_train, y_train)
```

```python
y_pred = clf1.predict(X_val)
acc = accuracy_score(y_val, y_pred, normalize=True)
misclass_err = 1 - acc
print("Accuracy:", acc, ", Misclassification error:", misclass_err)
```

```python
training_predicted = clf1.predict(X_train)
print(f"Classification report for classifier {clf1}:\n"
f"{metrics.classification_report(y_train, training_predicted)}\n")
```

```python
print('Confusion Matrix')
print(confusion_matrix(training_predicted, y_train))
```

```python
test_predicted = clf1.predict(X_val)
print('Confusion Matrix')
print(confusion_matrix(test_predicted, y_val))
print(f"Classification report for classifier {clf1}:\n"
f"{metrics.classification_report(y_val, test_predicted)}\n")
```

TP = 19. This shows that Logistic Regression shows better result in predicting the purchase.
