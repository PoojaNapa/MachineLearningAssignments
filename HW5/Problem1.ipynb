**(c) Plot Gini Index, Classification and Cross Entropy**

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, num=6)

gini= x*(1-x)
classi_err = 1 - np.maximum(x,1-x)
logx = np.zeros(6)
log_x = np.zeros(6)
for i in range(1,6):
    logx[i] = math.log(x[i])
for i in range(0,5):
    log_x[i] = math.log(1-x[i])
centropy = -((x*logx)+(1-x)*log_x)

plt.plot(x, gini, 'r--', label = 'Gini Index')
plt.plot(x, classi_err, 'b--', label = 'Classification Error')
plt.plot(x, centropy, 'g--', label = 'Cross Entropy')
plt.xlabel('p^m')
plt.ylabel('value')
plt.legend(bbox_to_anchor=(1.5, 1.0))
plt.show()
```
